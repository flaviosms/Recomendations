```python
import numpy as np
```

```python
class Input:
  def __init__(self, File):
    inputs=File.read().split()
    inputs=[int(i) for i in inputs]
    self.v=inputs[0]
    self.u=inputs[1]
    
    self.a_matrix=np.zeros((self.v,self.u)) #Initializing adjacency matrix
    
    self.v_list=[[] for i in range(self.v)] #Initializing adjacency lists
    self.u_list=[[] for i in range(self.u)]
    
    self.s_v_list=np.zeros((self.v,self.v)) #Initializing second order adjacency lists
    self.s_u_list=np.zeros((self.u,self.u))
    
    for i,x in enumerate(inputs[2:]):
        if(x==0):
            continue
        current_v=int(i/self.u)
        current_u=i%self.u
        self.v_list[current_v].append(current_u)
        self.u_list[current_u].append(current_v)
        self.a_matrix[current_v][current_u]=1
    for a,b in enumerate(self.v_list):
        for c in b:
            for d in self.u_list[c]:
                if(self.s_v_list[a][d]==0):
                    self.s_v_list[a][d]=1
                    self.s_v_list[d][a]=1
    for a,b in enumerate(self.u_list):
        for c in b:
            for d in self.v_list[c]:
                if(self.s_u_list[a][d]==0):
                    self.s_u_list[a][d]=1
                    self.s_u_list[d][a]=1
class Bicluster:
    def __init__(self, v_array,u_array):
        self.v_list=v_array
        self.u_list=u_array
        
class Solution:
    
    def __init__(self,entrada):
        
        self.a_matrix=np.zeros((entrada.v,entrada.u)) 
        self.v_clusters=self.greedy(entrada)
        for a in self.v_clusters:
            for b in a.v_list:
                for c in a.u_list:
                    self.a_matrix[b][c]=1
        
    def evaluate(self,e_a_matrix):
        npm1=e_a_matrix
        npm2=self.a_matrix
        return abs(npm1-npm2).sum()
        #return abs(np.matrix((npm1==npm2)-1).sum())
        
    def recomendations(self,e_a_matrix):
        npm1=np.matrix(e_a_matrix)
        npm2=np.matrix(self.a_matrix)
        rec=(npm1==npm2-1)
        rec_list=[]
        for i,x in enumerate(rec):
            for j,y in enumerate(x.flat):
                if(y):
                    rec_list.append([i,j])
        return rec_list
    
    def greedy(self,Input):
        
        g_v_list=Input.v_list.copy()
        g_u_list=Input.u_list.copy()
        while(g_v_list and g_u_list):
            max_value=0
            max_id=[]
            current_value=0
            current_id=[0,0]
            biclusters=[]
            for i,x in enumerate(g_v_list):
                for j,y in enumerate(g_u_list):
                    current_value=(Input.a_matrix[i][j]-0.5)*2 # w formula
                    v_links=len(x)
                    u_links=len(y)
                    for k,z in enumerate(Input.s_u_list[j]): # i Part of the in and diff formula
                        if(k==j):
                            continue
                        if(k in x):
                            if(z==0):
                                z=-1
                            else:
                                v_links-=1
                            current_value+=z

                    for k,z in enumerate(Input.s_v_list[i]): # j part of the in and diff formula
                        if(k==i):
                            continue
                        if(k in y):
                            if(z==0):
                                z=-1
                            else:
                                u_links-=1
                            current_value+=z
                    if(u_links<0 or v_links<0):
                        print("Error in number of links")
                        return "Error in number of links"
                    current_value-=u_links+v_links
                    if(max_id==[]):
                        max_id=[i,j]
                        max_value=current_value
                    elif(max_value<current_value):
                        max_id=[i,j]
                        max_value=current_value
            #Here the best candidate is choosen
            v_in_cluster=g_u_list[max_id[1]]
            u_in_cluster=g_v_list[max_id[0]]
            biclusters.append(Bicluster(v_in_cluster,u_in_cluster))
            newulist=[j for i, j in enumerate(g_u_list) if i not in u_in_cluster]
            newvlist=[j for i, j in enumerate(g_v_list) if i not in v_in_cluster]
            for i in v_in_cluster:
                for j in newulist:
                    if(i in j):
                        j.remove(i)
            for i in u_in_cluster:
                for j in newvlist:
                    if(i in j):
                        j.remove(i)
            print(newvlist,newulist,sep='\t')
            print(len(newvlist),len(newulist),sep='\t')
            g_v_list,g_u_list=newvlist,newulist
            
        if(g_v_list):
            for x in g_v_list:
                biclusters.append(Bicluster([],[x]))
        elif(g_u_list):
            for x in g_u_list:
                biclusters.append(Bicluster([x],[]))
        return biclusters
            
    
```

```python
f = open("V_5_U_7_D_0.5_S_1.dat",'r') 
#f = open("teste.dat",'r') 
entrada=Input(f)

#print("############")
#ex_biclusters=[Bicluster([0,2,3],[3,4,5,6]),Bicluster([1,4],[0,1,2])]
solucao=Solution(entrada)
#print(*entrada.s_v_list,sep='\n')
#print("############")
#print(*entrada.a_matrix,sep='\n')

#print(solucao.evaluate(entrada.a_matrix))
#print(entrada.u_list)
solucao.a_matrix
```
